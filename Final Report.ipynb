## Markup Ratio Investigation and Trend Identification

This analysis focuses on differences in Market Price over the 4 regions provided in the analysis. 

In order to understand these tends, we calculated the Market Price by taking the Retail Price for each city and dividing by the Farm Cost. This tells us the Markup Ratio, or how much higher the Retail Price for each city is over the Farm Price. 

For this analysis, we took the data segmented by product and grouped it by month year and unpivoted it so each city was in a single column. This was completed through the creation of a function and allowed us to create a function to quickly group and pivot the data. Another function was created to build each chart.

### Broccoli Market Price Analysis

Overall, broccoli shows a very tight City Retail to Farm Cost ratio. However, we can see that there is a large anomaly for the New York City price / Farm Price in February 2018. 



### Carrot Markup Price Analysis

The carrot Markup Price showed no obvious trends and the ratio varied grately between cities and years.



### Romaine and Cauliflower Markup Price Analysis

Romaine and cauliflower showed a close trend across all regions and each year, following a seasonal trend.

Strawberries also showed a trend, however it was not as consisent as the other products.






### Markup Ratio Analysis

The intra month variation of the Market Ratio for each Product was calculated by taking the Max Markup Ratio - the Min Markup Ratio. e.g. If the highest price for Cauliflower was in July 2016 in NY that was subtracted by the lowest price for July 2016, Atlanta.

From this the highest, lowest, average and standard devation for each year were calculated - to show how the in month variance changes.
